# Automating the _Digitale Charterbank Nederland_

The _Digitale Charterbank_ (DCN) Nederland is an online database of charters in Dutch archives, that strives to give access all charters stille available. DCN was made by Huygens ING, KNAW, in cooperation with De Ree, a company specializing in archive automation services. Most of DCN holdings are from their MAISFlexis database. 

De Ree provides the web service for their database, but it has no web api. For generating overviews of DCN it is highly desirable to have some automated access. For this we use web scraping using Python libraries, of which the requests library, Beautiful Soup and Pandas are most important.

See https://2.python-requests.org/en/master/, https://www.crummy.com/software/BeautifulSoup/bs4/doc/ and https://pandas.pydata.org/ for downloads/installation and documentation of these indispensable libraries. 

```python
#import necessary libraries
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
import html5lib # used for parsing
from urllib.parse import urlparse, parse_qs 
from collections import Counter
import matplotlib.pyplot as plt
```

### querying DCN

Now we are ready to query the charterbank and make an overview of the number of charters, the simplest task. We will use two functions for that. 

We want to query single years, but the web interface of DCN does not allow that, but equalling begin year and end year works around this restriction. The resulting number of result is on the result page in a HTML-element with class "mi_hits_hits_count". We concatenate all results and transform the list into a dataframe using the yr2dataframe function.

Note that for a large number of years this will take a considerable time as each page requires a separate request (in my case 1 min 38 s system time, Wall time 17+ minutes, all in all a long wait). It also hammers the DCN server, so use it cautiously.

```python
def get_years(url=url, begin=1000, end=1850, step=1):
    """collect the number of results from a range of years between
    begin and end with a step"""
    year_overz = []
    years = range(begin, end, step)
    pairs = [(years[i], years[i]+step-1) for i in range(len(years))]  # this is a hack but allows for any step size
    for pair in pairs:
        queryparts['mibj'] = pair[0]
        queryparts['miej'] = pair[1]   
        result = requests.get(url, params=queryparts)
        soup = BeautifulSoup(result.text, 'html5lib')
        number = soup.find_all(class_='mi_hits_hits_count')[0].text.strip() # there will only be one
        year_overz.append({'yr': '%s' % (pair[0]), 'number': number}) 
    return year_overz
    
def yr2dataframe(year_overz):
    yrdf = pd.DataFrame().from_records(year_overz)
    yrdf.number = yrdf.number.str.replace('.','')
    yrdf.number = yrdf.number.astype('int')
    return yrdf
```

```python
%%time

year_overz = get_years(url=url, begin=1000, end=1850, step=1 )
```

```python
yrdf = yr2dataframe(year_overz)
yrdf.head() #note that for the early years the numbers are very small
```

```python
# we need numeric values for plotting
yrdf['byr'] = yrdf['yr'].astype('int', inplace=True)
yrdf['nnr'] = yrdf.number.astype(int)
```

### Plot a graph
Jupyter makes (inline) plotting easy using matplotlib

```python

import matplotlib.ticker as ticker # for x-axis tick placement

x = yrdf.nnr 
y = yrdf.byr

fig = plt.figure(figsize=(20,20))
fig.suptitle("charters 1000-1850",fontsize=20)
ax1 = fig.add_subplot(111)
ax1.bar(y, x, color="blue")
ax1.xaxis.set_major_locator(ticker.MaxNLocator(nbins=20))
ax1.set_xlabel('year')
plt.show()
```

## Subject overview from a sample

### Handling query  output 

Now we want an overview of the subjects from DCN. This requires a few steps:

   1. a  output for sample years. We decided to make a systematic sample of each 10th year and group them by half century.
   2. get the data from the page and convert the descriptions to words
   3. make word clusters for some important subjects
    
Note that it is the intention to get a rough overview of important samples, so we only clean up results marginally

```python
# a function for cleaning up dataframes that tend to read extra columns from csv files
def clean(df):
    """clean empty values from dataframe"""
    xtdf = df.loc[~df.Titel.isna()]
    xtdf = xtdf.drop([column for column in xtdf.columns if 'Unnamed' in column], axis=1)
    return xtdf
```

```python
def df_from_soup(soup):
    """make a dataframe from charter soup"""
    tables = soup.find_all(class_="mi_table_view") # the css class contains the table content we need
    table = tables[0]
    df = pd.read_html(str(table))
    ndf = df[0]
    cdf = clean(ndf)
    return cdf
```

```python
# Pull the sample
res = {} # we make a list of dataframes first and concat afterwards
for item in range(925, 1850, 10):
    queryparts['mibj'] = item
    queryparts['miej'] = item+1    
    result = requests.get(url, params=queryparts)
    soup = BeautifulSoup(result.text, 'html5lib')
    df = df_from_soup(soup)
    try:
        res['%s' % item] = df
    except IndexError:
        pass
    
```

```python
#We did this earlier for reference so that we did not have to hit the charterportal with queries each time, just like the following cell

tdf = pd.concat(res.values(), ignore_index=True)`
tdf.to_csv('dcn/charter_csvs/totaal_overzicht.csv', sep='\t')
```

```python
import numpy as np
terms = [term for term in overzicht.term if type(term) == str]
terms
```

```python
# After inspection we came up with the following general terms and patterns

verzameltermen = {'archieftermen': ['chart.*', '^a[ck]+te.*', '^nummer', '\b(nr)', 'inventarisn(umme)?r'],
                 'koop' : ['ko[ch|g]+t', 'koop'],
                 'leen': ['(be)?leen', 'lijftocht'],
                 'pacht': ['.*pa[g|ch]+t.*'],
                 'land': ['^land\s+', 'morgen.*', '^akker', '.*grond.*'],
                 'rente': ['.*rente.*'],
                 'kerk': ['.*bisschop.*', '.*kapit.*', '.*kerk.*', '^tiend.*', 
                          'kloos.*', 'ab[td].*', 'altaa?r.*'],
                 'stad': ['.*st[a|e]d.*'],
                 'huis': ['hui[z|s]*'],
                 'benoeming': ['noeming.*', 'aanstelling']}
```

```python
norm = {}
for verzamelterm in verzameltermen.keys():
    zktermen = []
    for pat in verzameltermen[verzamelterm]:
        zktermen.extend([t for t in terms if re.search(pat, t) != None])
    norm[verzamelterm] = zktermen
```

```python
#manually adjust this for off terms. We still need to do this for the rest
norm['leen'] =['beleend',
 'beleening',
 'beleent',
 'borgleen',
 'leen',
 'leen-',
 'leen-en',
 'leenakte',
 'leenbrief',
 'leendert',
 'leengoed',
 'leengoederen',
 'leenhof',
 'leenhulde',
 'leenkamer',
 'leenmannen',
 'leenroerig',
 'leenten',
 'leenten:1/4',
 'leentyen',
 'ringie-leen',
 'sleen',]
```

```python

# this is necessary for categorizing the values in the dataframe

rev_norms = {svalue: key for key, value in norm.items() for svalue in value}
rev_norms

# the resulting list is generally ok, and the terms that are off have very small incidence, so we do not clean it manually
```

```python
overzicht['normterm'] = overzicht.term.replace(rev_norms)
overzicht.normterm = overzicht.normterm.astype('category')
for term in verzameltermen:
    print(term, overzicht.query("normterm == '{}'".format(term)).total.sum())
```
